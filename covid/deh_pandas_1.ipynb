```python
import pandas as pd
from warnings import filterwarnings
import numpy as np

filterwarnings('ignore')
```

```python
lista_acoes = []

for posicao in range(30):
    lista_acoes.append('acao_' + str(posicao))

lista_acoes
```

```python
s_acoes = pd.Series(lista_acoes)
s_acoes
```

```python
s_acoes.iloc[10]
```

```python
np.random.seed(10)
s_preco = pd.Series(np.random.uniform(20,50,30).round(2))
s_preco
```

```python
s_dividendo = pd.Series(np.random.uniform(0.5,5,30).round(2))
s_dividendo
```

```python
dados = {
    'ticker': s_acoes,
    'preco': s_preco,
    'dividendo': s_dividendo
}

df = pd.DataFrame(dados)
df.head()
```
